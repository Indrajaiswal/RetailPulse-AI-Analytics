# RetailPulse_AI_Analytics

# Dataset Overview

The Online Retail II dataset contains transactional sales records from a UK-based online retail company between December 2009 and December 2011.

The company primarily sells unique gift products, and many customers are wholesalers. The dataset includes customer purchases, product details, transaction timestamps, quantities, pricing information, and customer geographic information.

This dataset is highly suitable for retail analytics because it enables the analysis of customer purchasing behavior, sales trends, product demand, and customer value patterns.

The primary business problems addressed using this dataset include:

- Understanding customer purchasing behavior
- Identifying high-value and loyal customers
- Forecasting future sales demand
- Detecting customer churn risk
- Optimizing inventory management
- Improving business decision-making through data-driven insights

# Problem-Solving
The Quantity feature helps analyze product demand patterns and supports inventory optimization and demand forecasting.

# Data Collection:
The dataset has been referred from: 
https://www.kaggle.com/datasets/lakshmi25npathi/online-retail-dataset

| Column      | Description                    | Business Use                               |
| ----------- | ------------------------------ | ------------------------------------------ |
| Invoice     | Unique transaction/bill number | Identify and group customer transactions   |
| StockCode   | Unique product identifier      | Product-level sales analysis               |
| Description | Product name/details           | Product insights and demand analysis       |
| Quantity    | Number of units purchased      | Demand forecasting and inventory analysis  |
| InvoiceDate | Transaction date and time      | Time-series analysis and sales forecasting |
| Price       | Price per product unit         | Revenue and pricing analysis               |
| Customer ID | Unique customer identifier     | Customer segmentation and churn analysis   |
| Country     | Customer geographic location   | Regional and market analysis               |


```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = ("../data/rawdata/online_retail_II.xlsx")

xls = pd.ExcelFile(file_path)

df_2009 = pd.read_excel(xls, sheet_name="Year 2009-2010")
df_2010 = pd.read_excel(xls, sheet_name="Year 2010-2011")

df = pd.concat([df_2009, df_2010], ignore_index=True)
```

```python
df = df.dropna(subset=['Customer ID'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df[df['Invoice'].astype(str).str.startswith('C')].shape
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df[df['Quantity'] <= 0].shape
```

```python
df = df[df['Quantity'] > 0]
```

```python
df[df['Price'] <= 0].shape
```

```python
df = df[df['Price'] > 0]
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Remove non-product administrative records

exclude_keywords = [
    "Manual",
    "POSTAGE",
    "BANK CHARGES",
    "ADJUST"
]

df = df[~df["Description"].isin(exclude_keywords)]
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
df['Weekday'] = df['InvoiceDate'].dt.day_name()
```

```python
df.to_csv("../data/processed/cleaned_retail_data.csv", index=False)
```
